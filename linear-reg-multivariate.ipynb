```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn import linear_model
```

```python

df=pd.read_csv("homepri.csv")
df
```

```python
#3000 sqr fr area ,3 bedroom 40yrs
#2500 sq fit area,4bedrom ,5 yrs
          #price = m1*area+m2*bedroom +m3*age+b      (b=incetpect)
```

```python
#data cleaning
import math

median_bed=math.floor(df.bedrooms.median())
median_bed
```

```python
df.bedrooms=df.bedrooms.fillna(median_bed)        #filling the median in place of null
df
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)          
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
reg.predict([[2500,4,5]])
```
